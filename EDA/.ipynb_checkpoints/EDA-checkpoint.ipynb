```python
df = pd.read_csv("car_price.csv")
df
```

```python
sorted_df = df.sort_values('0-60 MPH Time (seconds)')
sorted_df.head(10)
```

```python
def energy_filter(df,energy_type='all'):
    if energy_type=='all':
        return df
    if energy_type=='eletric':
        electric_cars_df = df[df['Engine Size (L)'].astype(str).str.contains('Electric')]
        return electric_cars_df
    if energy_type=='fuel':
        filtered_df=df[df['Engine Size (L)'].notnull() & df['Engine Size (L)'].apply(lambda x: isinstance(x, (int, float)))]
        return filtered_df
    
    
```

```python
energy_type='all'
# energy_type='fuel'
# energy_type='eletric'
energy_filter(sorted_df,energy_type)
```

```python
import pandas as pd
import numpy as np

def filter_by_0_to_60(df, l, r):
    """
    Filter the DataFrame based on the 0-60 MPH Time (seconds) column
    
    Parameters:
    - df: The input DataFrame
    - l: The lower bound of the 0-60 MPH Time (seconds) values
    - r: The upper bound of the 0-60 MPH Time (seconds) values
    
    Returns:
    A new DataFrame that includes only the rows with 0-60 MPH Time (seconds) values between l and r
    """

    df['0-60 MPH Time (seconds)'] = pd.to_numeric(df['0-60 MPH Time (seconds)'], errors='coerce')

    filtered_df = df[(df['0-60 MPH Time (seconds)'] >= l) & (df['0-60 MPH Time (seconds)'] <= r)]
    
    return filtered_df
```

```python
filter_by_0_to_60(df, 0, 1.9)
```

```python
import pandas as pd

def filter_by_model(df, model):
    """
    Filter the DataFrame based on the Car Model column
    
    Parameters:
    - df: The input DataFrame
    - model: The specific Car Model to filter
    
    Returns:
    A new DataFrame that includes only the rows with Car Model value equals to the specified model
    """
    filtered_df = df[df['Car Model'] == model]
    return filtered_df
```

```python
filter_by_model(df, 'Mustang')
```
